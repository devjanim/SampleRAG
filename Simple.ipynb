```python
import os
mykey = os.getenv("OPENAI_APIKEY")
llm = OpenAI(api_key=mykey)
response=llm.chat.completions.create(
    model="gpt-5.2",
    messages=[{"role": "user", "content": "What is the capital of France?"}]
)
print(response.choices[0].message.content)
```
